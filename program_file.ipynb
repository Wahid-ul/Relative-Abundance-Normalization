```python
#importing library
import pandas as pd
```

```python
#task1-input file
data=pd.read_csv("overall_KEGG_func_abund.tsv",sep="\t")
```

```python
data
```

```python
#dropping those rows having Unclassified,Unmapped ID
data.drop([3476,3477],inplace=True)
```

```python
#Relative abundance calculation of control sample
RA_Control=(data.iloc[:,1]/data.iloc[:,1].sum())*100
RA_Control
```

```python
#Relative abundance calculation of control sample
RA_Test=(data.iloc[:,2]/data.iloc[:,2].sum())*100
RA_Test
```

```python
#creating Relative abundance column for both control and test
data['RA_control']=RA_Control
data['RA_test']=RA_Test
```

```python
#task2
df=pd.read_csv("overall_KEGG_names.tsv",sep="\t")
df
```

```python
df1=pd.read_csv("overall_KEGG_pths.tsv",sep="\t")
df1
```

```python
#since we have to take common value from coulmn based on ID so we have 
#to keep same name as ID for three datasets
data.rename(columns={'ID':'KO_Id'},inplace=True)
data
```
